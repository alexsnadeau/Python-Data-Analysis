```python
#now dropping all NaN values
#DataFrame.dropna(*, axis=0 [0 for dropping rows which contain missing values, 1 for dropping columns which contain missing value], how=_NoDefault.no_default, thresh=_NoDefault.no_default, subset=None [column label], inplace=False, ignore_index=False)

df.dropna(subset=["salary_year_avg"], inplace=True)
```

```python
#data analysis

import pandas as pd
from datasets import load_dataset

#Loading Data
dataset=load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()

#Data Cleanup
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

```python
#idxmin gets the index or row to where the minimum value is

min_salary = df["salary_year_avg"].idxmin()
```

```python
#calling the variable min_salary because it is the index or row at which the minimum value is stored in

df.iloc[min_salary]
```

```python
#counts the number of strings it finds in the column, just like a normal count function

df.job_title_short.value_counts()
```

```python
#groupby is used to make dataframes with agg functions such as sum, median, mean, count, size, etc
#df.groupby("columns to group")["columns to aggregate"].your_agg_method
#this groupby makes the index job_title_short and looks at the median salary_year_avg values for each job title

df.groupby("job_title_short")["salary_year_avg"].median()
```

```python
#if we want to groupby multiple columns, we need to make it a list and it applies the agg function to both columns together

df.groupby(["job_title_short", "job_country"])["salary_year_avg"].median()
```

```python
#if we want to aggregate on multiple columns, then make the second part a list and it will apply the agg function to different columns

df.groupby("job_title_short")[["salary_year_avg", "salary_hour_avg"]].median()
```

```python
#want to do multiple operations on the groupby, such as max and min
#use .agg("[operation list]")

df.groupby("job_title_short")["salary_year_avg"].agg(["min", "max", "median"])
```
